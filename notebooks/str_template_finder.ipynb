```python
def generate_simple_config(seq):
    def p(x):
        return x[0]
    def d(x):
        if len(x) > 1 and x[1]:
            return True
        return False
    def e(x):
        if len(x) > 2:
            return x[2]
        return 0

    return {
        'template_structure': [
            {'pattern': p(x),
             'delim': d(x),
             'max_errors': e(x) 
            }
            for x in seq
        ]
    }
```

# Template generation

```python
print("\n".join(str_markers.locusName))
```

```python
from collections import defaultdict, Counter

MIN_BASE_REPEATS = 5

for ref_i, ref in enumerate(str_markers.locusReferenceSequence):
    print('\n\n\n', str_markers.iloc[ref_i].locusName, '\n', ref)
    
    # Base repeat length
    base_repeats = []
    for k in range(2, 7):
        repeats = Counter()
        for i in range(len(ref)):
            if i+k >= len(ref):
                continue
            repeats.update([ref[i:i+k]])
    
        for repeat, occ in repeats.most_common(5):
            if repeat * MIN_BASE_REPEATS in ref:
                base_repeats.append(repeat)
        
    print(str_markers.iloc[ref_i].repeatPattern)
    print(base_repeats)
```
